```python
pop_anime['token'].apply(lambda x:x.replace(' ',''))
```

```python
pop_anime.token=pop_anime['token'].apply(lambda x: ' '.join(x))
```

```python
pop_anime.token=pop_anime['token'].apply(lambda x: x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
pop_anime.token=pop_anime['token'].apply(stem)
```

```python
pop_anime
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000,stop_words='english')
```

```python
vectors=(cv.fit_transform(pop_anime.token)).toarray()
```

```python
vectors.shape
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
pop_anime.to_csv('pop_anime.csv',index=False)
```

```python
pop_anime
```

```python
import pickle
pickle.dump(similarity,open('sim.sav','wb'))
```

```python
def recommend(anime):
    anime_index=pop_anime[pop_anime['name']==anime].index[0]
    distances=sorted(list(enumerate(similarity[anime_index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(pop_anime.iloc[i[0]]['name'])
    
```

```python
recommend('Black Clover')
```
